```python
#lib(s)
import matplotlib.pyplot as plt
import random
import os
import cv2
import math
import numpy as np
#hand detector module     it uses mediapipe
from cvzone.HandTrackingModule import HandDetector
```

```python
#using cv2 open webcam
cap=cv2.VideoCapture(0)
#hand detector
detector=HandDetector(maxHands=1)
#i to give a distinct file name
i=0
#offset is to add some extra pixels over hand
offset=20

folder='E:\Signdataset'
while True:
    #read the image from webcam
    success,img=cap.read()
    #detect the hand
    hands,img=detector.findHands(img)
    if hands:
        #if there is any hand take the 1st hand
        hand=hands[0]
        #dimensions of the hand
        x,y,w,h=hand['bbox']
        #cropped image which contain only hand
        imgCrop=img[y-offset:y+h+offset,x-offset:x+w+offset]
        #to show the cropped image
        cv2.imshow("Imgae Crop",imgCrop)
    #to show the whole image
    cv2.imshow("Imgae",img)
    #wait for any key to be pressed
    key=cv2.waitKey(1)
    #if key is s: save in the distict file
    if key==ord('s'):
        i+=1
        cv2.imwrite(f"{folder}/15/back_"+str(i)+".jpg",imgCrop)
```
